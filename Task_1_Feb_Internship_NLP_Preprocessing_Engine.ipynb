## Task 1: Conceptual Understanding

```python
1.Difference between "Love" and "love" in NLP
"Love" vs "love" differs only in case. In NLP, most preprocessing converts text to lowercase so that 
"Love" and "love" are treated as the same token. Without lowercasing, NLP models may consider them 
different words, which can take duplicates.
2.What happens if stopwords are not removed
Stopwords (like “the”, “is”, “at”) are very common words that carry little semantic meaning.
If they are not removed, they can dominate token counts, skew frequency analysis, and add
noise to text-based models like topic modeling or sentiment analysis.
3.Two real-world scenarios where removing stopwords can be harmful
Sentiment analysis: Words like “not” or “no” are stopwords but crucial for meaning.
Example: "I am not happy" → removing “not” changes sentiment.
Question-answering / search engines: Words like “what”, “who” might be stopwords but 
are essential for the query.
Example: "Who is the CEO of OpenAI?" → removing “who” loses intent.
4.Difference between stemming and lemmatization
Stemming: Cuts words to their root form using simple rules (may produce non-dictionary words).
Example: "running" → "run", "better" → "better" (might not change).
Lemmatization: Converts words to their dictionary base form using vocabulary and context.
Example: "running" → "run", "better" → "good".
Key: Lemmatization is more accurate; stemming is faster but crude.
```

## Task 2:Build Advanced Preprocessing Function

```python
import re  #re is regualr exprestion userd for search and matching
import string #it is library with predefined sstring consonanys,letter and all
import nltk

from nltk.corpus import stopwords #corpus means colelction of the text
from nltk.stem import WordNetLemmatizer #stem means reducing words to its stem

nltk.download('stopwords')#downloads the dataset
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english')) # unique stopwords
lemmatizer = WordNetLemmatizer()
```

## Task 3: Stress Testing


```python
def preprocess_text(text):
    #1 .remove number
    text = re.sub(r'\d+', '', text)

    # 2. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

     #3. Handle repeated characters (more than 2 repeats → reduce to 1)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    #4. Lowercasing
    text = text.lower()
    
    #.5Remove punctuations
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    #6.Remove Special Characters/Emojis
    text = re.sub(r'[^\w\s]', '', text)
    
    #7. Tokenization
    tokens = text.split()
    
    #8. Stopwords removal
    tokens = [word for word in tokens if word not in stop_words]

    return tokens
```

```python
# Test input: 10 diverse sentences
text = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Stress test
for i, text in enumerate(text, 1):
    tokens = preprocess_text(text)
    clean_sentence = ' '.join(tokens)
    print(f"text {i}:")
    print("Original Text:",text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    
```

## Task 4: Token Analytics


```python
def token_stats(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = round(sum(len(t) for t in tokens) / total_tokens, 2) if total_tokens > 0 else 0
    return total_tokens, unique_tokens, avg_token_length
text = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for i, text in enumerate(text, 1):
    tokens = preprocess_text(text)
    clean_sentence = ' '.join(tokens)
    total, unique, avg_len = token_stats(tokens)
    
    print(f"text {i}:")
    print("Original Text:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", avg_len)
    
```

## Task 5: Frequency Analysis


```python
all_tokens = [
    # Text 1
    "get", "free", "access", "now",
    # Text 2
    "absolutely", "looooved", "product",
    # Text 3
    "worst", "service", "ever",
    # Text 4
    # empty or phone removed → no token
    # Text 5
    "best", "course",
    # Text 6
    "visit", "now",
    # Text 7
    "no",  # assuming 'nooooo' reduced to 'no'
    "bad",
    # Text 8
    "ok", "ok", "ok", "got", "it",
    # Text 9
    "win", "now", "limited", "offer",
    # Text 10
    "happy"
]

from collections import Counter

# Count frequency of all tokens
token_counts = Counter(all_tokens)

# Top 10 most frequent words
top_10 = token_counts.most_common(10)

# Top 5 least frequent words
# Least frequent = frequency = 1
least_5 = [word for word, count in token_counts.items() if count == 1][:5]

print("Top 10 most frequent words:", top_10)
print("Top 5 least frequent words:", least_5)
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(' '.join(tokens))
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Example usage:
texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

result = full_pipeline(texts)
print(result)
```
