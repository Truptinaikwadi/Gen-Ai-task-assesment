```python
#Step 2: Import Libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
#step 3 Choose a model
model_name = "microsoft/DialoGPT-small"  # Small DialoGPT, good for chat

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Step 3: Print initial greeting
print("Chatbot: Hello! I am your AI assistant. How can I help you today?")

# Step 4: Initialize chat history to remember conversation context
chat_history_ids = None

# Step 5: Start conversation loop
while True:
    # Get user input
    user_input = input("You: ")
    user_lower = user_input.lower()  # Convert to lowercase for easy matching

    # Step 6: Exit condition
    if user_lower in ["exit", "quit"]:
        print("Chatbot: Goodbye!")
        break  # Exit the loop and end program

    # Step 7: Hardcoded responses for specific keywords
    if user_lower == "hello":
        print("Chatbot: Hello! Nice to meet you. How can I assist you today?")

    elif "artificial intelligence" in user_lower:
        print("Chatbot: Artificial Intelligence refers to the simulation of human intelligence by machines that can perform tasks such as learning, reasoning, and problem solving.")

    elif "python" in user_lower:
        print("Chatbot: Python was created by Guido van Rossum and released in 1991.")

    elif "thank" in user_lower:
        print("Chatbot: You're welcome! Feel free to ask more questions.")

    # Step 8: Fallback to AI-generated response for all other inputs
    else:
        # Encode user input + end-of-sequence token
        new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

        # Concatenate with chat history if it exists
        if chat_history_ids is not None:
            input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
        else:
            input_ids = new_input_ids

        # Attention mask: all ones (all tokens are valid)
        attention_mask = torch.ones(input_ids.shape, dtype=torch.long)

        # Generate response from the model
        chat_history_ids = model.generate(
            input_ids,
            attention_mask=attention_mask,
            max_length=500,           # Maximum response length
            pad_token_id=tokenizer.eos_token_id,
            do_sample=True,           # Enable randomness for varied replies
            top_k=50,                 # Consider top 50 probable tokens
            top_p=0.9,                # Nucleus sampling
            temperature=0.7           # Creativity control: 0.7 = moderate
        )

        # Decode the model's output to readable text
        response = tokenizer.decode(chat_history_ids[:, input_ids.shape[-1]:][0], skip_special_tokens=True)

        # Print the chatbot's response
        print(f"Chatbot: {response}")
```
